```python
import os
import cv2 as cv
import numpy as np
from keras.models import load_model
```

```python
model = load_model('CovidModel.h5')
```

```python
categories = os.listdir('Training/')
categories
```

```python
import matplotlib.pyplot as plt
img = cv.imread('Training/normal/IM-0011-0001-0002.jpeg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img = cv.resize(img, (224, 224))
print(img.shape)

plt.imshow(img)
plt.show()

pred = model.predict(np.expand_dims(img/255, axis=0))
pred = np.argmax(pred)
result = categories[pred]
print(result)
```
